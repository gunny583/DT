# Chapter 2. Causality and Experiments 

- 인과관계<span style="color:red"> (Causality: Cause and Effect) </span>
 - 원인과 결과를 규명
   - 사형제도가 범죄율을 낮추는가?
   - 초콜릿은 건강에 유익한가?
   - 암을 일으키는 원인은 무엇인가?
 - ___상관관계(correlation, association)___ 와 비슷하나 구분되어야 한다.
- Data science를 통해 인과관계를 설명하는 것은 가장 중요한 목적이다.


- Observational study
  - _treatment_ : 인자(factor, variable)
  - _outcome_ : 영향결과(effect)
- 초콜릿은 건강에 유익한가?
  - _treatment_ : 초콜릿을 먹은 사람들과 그렇지 않은 사람들
  - _outcome_ : 건강지표 (혈압)
- 사형제도가 범죄율을 낮추는가?
  - _treatment_ : 사형제를 시행하는 국가들과 그렇지 않은 국가들
  - _outcome_ : 국가별 범죄율 (10만명당 살인범죄건수)
  
- 일반적으로 두 단계의 분석이 필요하다.
    - Association : treatment와 outcome 사이에 상관성이 있다.
    - Causality : 그러한 상관성에는 인과관계가 있다.
  

### Observation and Visualization 
- Case study - John Snow and the Broad Street Pump
- 1850년대 런던
 - 질병, 특히 콜레라 빈번
 - 그 원인을 나쁜 공기(악취; miasmas)라고 생각함. (1883년 콜레라 발병 메커니즘 발견)
- John Snow는 그러한 원인에 의심
 - 어느 집 가족들이 모두 콜레라로 죽었으나 바로 옆 집은 영향을 받지 않음
 - 구토와 설사가 수반. 공기가 아니라 음식이 원인이라 생각.
 - 지도에 사망자 거주지를 표시해가며 추적.

John Snow의 [Map](https://en.wikipedia.org/wiki/John_Snow)

- black bar : 사망자 수
- black disc : 수도 펌프의 위치

- 분석
  - 대부분의 사망자는 Broad St. 펌프 주변
  - 일부 사망자들은 Rupert St. 펌프가 더 가까이 있지만 막힌 도로여서 실제로는 Broad St. 펌프를 사용
  - Broad St. 펌프 동쪽 두 블럭에서는 사망자가 없는데 이는 그 사이에 양조장이 있어서 그 곳 물을 사용했기 때문
  - Broad St. 펌프에서 멀리 떨어진 곳에서 사망자들이 있는데 이들은 모두 어린이였으며 학교에서 돌아오는 길에 Broad St. 펌프에서 물을 자주 마심.
  - 매우 멀리 떨어진 곳에 사는 사망자가 있었는데 조카가 Broad St. 펌프물을 자주 가져다 주었기 때문
- 이후 실제로 Broad St. 펌프가 누수로 하수물에 오염되었음이 밝혀짐

- 이러한 지도 분석을 통해 시청을 설득, Broad St. 펌프를 제거함. 
  - “Where is the handle to this pump?”

### Snow’s “Grand Experiment” 
- 1849년 콜레라
 - 두 수도공급회사 (Lambeth와 S&V) 모두 오염된 템즈강에서 취수 
 - 비슷한 사망자 비율
- 1853-54년 콜레라 (아래 표 참조)
 - Lambeth는 상류지역으로 취수원을 이동
 - S&V 공급지역의 사망자 비율이 월등히 높음

### Causality
- 수도공급회사가 콜레라 사망의 원인인가?
  - _treatment_ group (S&V 공급 가구) vs _control_ group (Lambeth 공급 가구)
  - _outcome_ : 사망자 비율
- Treatment그룹과 Control그룹 사이에 다른 차이점은 없는가?
  - _confounding factors_
  - 소득, 성별, 연령대, ...
- 커피와 폐암
  - 1960년대 많은 연구에서 커피를 마시는 사람들과 폐암발병률 사이에 연관성이 있음을 보임.
  - 커피가 폐암을 일으키는가?
  - _confounding factor_ : 흡연비율


### Randomization
- Confounding factor를 걸러내는 방법론 중의 하나
- Randomized controlled trial (RCT)
  - Treatment/Control 그룹을 전체에서 임의로(randomly) 선택하여 선정
  - 플라시보 효과(Placebo effect)-실제로는 control group이지만 마치 treatment group인 듯
- 사례연구) 복지보조금 지급 (멕시코)
  - 목적: 저소득층 가정내 아이들의 학교등록비율을 높이기 위해 보조금 지급 여부 검토
  - treatment group: 학교에 아이들이 다닐 결우 보조금 지급
  - control group: 나머지 가정의 아이들
  - 결과: 남학생의 경우 등록율이 73%(control group)에서 77%(treatment group)로, 여학생의 경우 67%에서 75%로 증가
  - 적용: 제도를 본격적으로 시행함 

### Recap
- Observation study : treatment vs control groups
  - From association(relation) to causal association
- Confounding factor
  - Randomized controlled trial
