## Stylesheets

자신만의 stylesheet를 만들거나 이미 포함된 것을 활용함으로써 style을 변경할 수 있다. 

````python
plt.style.available[:5]
````

다음과 같이 호출하면 그 이후로 계속 변경된 style이 적용된다.
``` python
plt.style.use('stylename')
```

만일 일부만 적용한다면 다음과 같이 사용하면 된다. 
``` python
with plt.style.context('stylename'):
    make_a_plot()
```


연습을 위해 histogram과 lines을 그리는 함수를 만든다.

````python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
````

### FiveThiryEight style

````python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
````

### ggplot


````python
with plt.style.context('ggplot'):
    hist_and_lines()
````

### Bayesian Methods for Hackers

````python
with plt.style.context('bmh'):
    hist_and_lines()
````

### Dark background



````python
with plt.style.context('dark_background'):
    hist_and_lines()
````

### Grayscale


````python
with plt.style.context('grayscale'):
    hist_and_lines()
````

### Seaborn style


````python
import seaborn
hist_and_lines()
````
