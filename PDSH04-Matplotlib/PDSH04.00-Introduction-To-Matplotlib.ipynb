### ``show()`` or No ``show()``? How to Display Your Plots

실행 환경이 다음의 경우에 따라 조금씩 차이가 있다. 
- script
- IPython terminal
- IPython notebook.

#### Plotting from a script

아래 예 *myplot.py* 와 같이 script를 만들면 된다. (``plt.show()``가 그림을 그리는 함수이다.)

```python
# ------- file: myplot.py ------
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

그리고 다음과 같이 실행하면 그림을 나타내는 window가 열린다. 

```
$ python myplot.py
```

주의)  ``plt.show()``은 그림 개수와 상관없이 Python session 내 한 번만 호출하면 된다. (따라서 script 맨 마지막에 보통 배치한다.)

#### Plotting from an IPython shell

Matplotlib mode를 지정하기 위해 ``%matplotlib`` magic command 를 실행하면 된다.

```ipython
In [1]: %matplotlib
Using matplotlib backend: TkAgg

In [2]: import matplotlib.pyplot as plt
```

이후  ``plt`` plot command가 발생하면 plot이 업데이트되며 혹은 ``plt.draw()``를 호출하여 명시적으로 업데이트한다. 

#### Plotting from an IPython notebook

IPython shell과 마찬가지로 IPython notebook에서도 ``%matplotlib`` command를 통해 실행한다. 
이 때 다음 2가지 option이 가능하다. 

- ``%matplotlib notebook``: *interactive* plot 생성
- ``%matplotlib inline``: *static* image 생성 

여기에서는 ``%matplotlib inline``을 사용한다.

````python
%matplotlib inline
````

이후 plot 결과는 PNG image를 생성한다. 

````python
import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
````

### Saving Figures to File

Matplotlib 장점 중 하나는 다양한 그림 포맷을 지원한다는 점이다. 
``savefig()`` command를 사용하면 된다. 

````python
fig.savefig('my_figure.png')
````

## Two Interfaces for the Price of One

Matplotlib에서 약간 혼란스럽게 되는 이유는 다음 2가지 interface 제공 때문이다.
- MATLAB-style state-based interface
- object-oriented interface.

#### MATLAB-style Interface



````python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
````

#### Object-oriented interface


````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
````

``plt.plot()``을 ``ax.plot()``로 바꾼 약간의 차이이지만 접근 방식이 좀 더 복잡하고 강력한 기능을 제공한다.
